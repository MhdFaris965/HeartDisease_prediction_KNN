```python
from sklearn.model_selection import train_test_split
```

```python
X=df.drop('HeartDisease',axis=1)
y=df['HeartDisease']
```

```python
#STANDARD NORMAL DISTIBUTION
#VARIANCE =1
#MEAN = 0
#scaling feature
from sklearn.preprocessing import StandardScaler
```

```python
for i in df.columns:
    print(i,df[i].mean())
```

```python
ss=StandardScaler()
X=ss.fit_transform(X)
```

```python
X=pd.DataFrame(data=X,columns=['Age ','Gender ','BloodPressure ','Cholesterol ','HeartRate ','QuantumPatternFeature '])
for i in X.columns:
    print(i,X[i].mean())
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn=KNeighborsClassifier(n_neighbors=7)
```

```python
knn.fit(X_train,y_train)
```

```python
y_pred_train=knn.predict(X_train)
y_pred_test=knn.predict(X_test)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_train,y_pred_train))
```

```python
print(classification_report(y_test,y_pred_test))
```

```python
error_rates = []#list all errors

for i in range(1,21): #for loop trying different values
    knn=KNeighborsClassifier(n_neighbors=i) # Making Model
    knn.fit(X_train,y_train) #Fitting Training Data
    y_pred_test=knn.predict(X_test) #predicting on testing Data
    error=np.mean(y_pred_test!=y_test) #calculating error
    error_rates.append(error) #adding that error to error_rates
```

```python
#ELBOW METHOD
plt.figure(figsize=(10,5),)
plt.plot(range(1,21),error_rates,marker='o',linestyle='--')
plt.title('Error Rate vs K value')
plt.xlabel('K value')
plt.ylabel('Error Rates')
```
